```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
def serie_saisonniere_logitrend(t,Capacity_max=1000, taux_croissance=0.02, decal_temporel=20,
                                period=12, amplitude_sais=100):
    C = Capacity_max
    k = taux_croissance
    t0 = decal_temporel
    T = period
    A = amplitude_sais
    epsil = np.random.randn(len(t))

    y = [(C / (1 + np.exp(-k*(t[i]-t0))) + A * np.sin((2* np.pi * t[i]) / T) + epsil[i]) for i in t]

    return y
```

```python
t = np.arange(0,1000)
y = serie_saisonniere_logitrend(t)

plt.plot(t,y)
```
