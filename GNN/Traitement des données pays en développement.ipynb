```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
```

```python
# LDR pour dire Less Development Regions

# L'adresse du fichier
file_path = 'Data/WPP2024_MORT_F06_1_SINGLE_AGE_LIFE_TABLE_ESTIMATES_BOTH_SEXES.xlsx'

# Les colonnes à sélectionner dans le fichier
selected_cols = ['Region, subregion, country or area *', 'Year', 'Age (x)', 'Central death rate m(x,n)', 
            'Survival ratio S(x,n)', 'Expectation of life e(x)']

# Lecture de la base
LDR_data = pd.read_excel(file_path,skiprows=16,usecols=selected_cols)

# Renommer les colonnes 
LDR_data = LDR_data.rename({'Region, subregion, country or area *': 'Region_Country',
                            'Age (x)': 'Age',
                            'Central death rate m(x,n)': 'death_rate'})
nigeria_index = LDR_data[]
```
